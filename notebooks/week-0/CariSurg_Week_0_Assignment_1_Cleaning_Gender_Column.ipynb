```python
# conditions

def classify_pulse(bpm):
  if bpm < 60:
    return "Bradycardia (too slow)"
  elif bpm > 100:
    return "Tachycardia (too fast)"
  else:
    return "Normal"

# testing
test_pulses = [59, 80, 150]
for p in test_pulses:
  print (f" Pulse {p} bpm -> {classify_pulse(p)}")
```

```python
# seeing what values are actually in the gender column
print("Unique values in Gender:")
print(df_raw['Gender'].unique())
print(f"\nTotal unique values used: {df_raw['Gender'].nunique()}")
```

```python
# conting how many of each we have
print("Gender Value counts: ")
print(df_raw['Gender'].value_counts())
```

```python
# grouping male and female variations, and assigning them to 1 and 0 for easier interpretation
gender_map = {
    'Male':   1,
    'MALE':   1,
    '1':      1,
    'Female': 0,
    'FEMALE': 0,
    '0':      0
}
```

```python
df_raw['Gender_Clean'] = df_raw['Gender'].map(gender_map)

print("After cleaning Gender Column:")
print(df_raw['Gender_Clean'].value_counts())

print(f"Data type: {df_raw['Gender'].dtype}")
print ("\n")
df_raw.head()
```
